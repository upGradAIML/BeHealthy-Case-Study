```python
# Question 3

def extract_features(doc, i):
    features = {}

    # Feature 1: Check if xi starts with a capital letter
    features['F1'] = doc[i].text.istitle()

    # Feature 2: Check if xi is the parent of 'nsubj' or 'nsubjpass' dependency tagged words
    features['F2'] = any(child.ent_type_ in ['E', 'P'] for child in doc[i].children if child.dep_ in ['nsubj', 'nsubjpass'])

    # Feature 3: Check if xi is the parent of 'dobj' dependency tagged words
    features['F3'] = any(child.ent_type_ in ['E', 'P'] for child in doc[i].children if child.dep_ == 'dobj')

    return features
```

```python
sent4 = 'Harry is not gardening as it is raining'
doc = nlp(sent4)
for i, token in enumerate(doc):
    features = extract_features(doc, i)
    print(f"Token: {token.text}, Features: {features}")
```
